```python
# Load DATA OND ATAFRAME& Explore Data

df = pd.read_csv("/content/data.csv")
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df.dropna(axis=1,inplace=True)
```

```python
df.shape
```

```python
df['diagnosis'].value_counts()
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
df.iloc[:,1]=labelencoder.fit_transform(df.iloc[:,1].values)
```

```python
df
```

```python
X=df.iloc[:,2:].values
Y=df.iloc[:,1].values
```

```python
#splitting datasets into training (75%) and tetsing (25%)
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25)
```

scaling data

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train= sc.fit_transform(X_train)
X_test= sc.transform(X_test)
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression()

Y_train = Y_train.astype(int)		#Y_train to numeric

Y_test = Y_test.astype(int)		        #Y_test to numeric

classifier.fit(X_train,Y_train)

classifier.get_params()
```

```python
#use of trained model to make predictions on test data
predictions= classifier.predict(X_test)
```

```python
#plot confusion matrix
from sklearn.metrics import confusion_matrix
import seaborn as sns
cm=confusion_matrix(Y_test,predictions)
print(cm)
sns.heatmap(cm,annot=True)
```

```python
#get accuracy score
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,predictions))
```
